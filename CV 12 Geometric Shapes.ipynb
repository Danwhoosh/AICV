## Write a program in python to draw different geometric shapes and to write text on images using OpenCV

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
img = np.zeros((1000, 1000, 3), dtype = "uint8")
```

```python
cv2.rectangle(img, (200, 400), (800, 1000), (255, 0, 255), 5)
cv2.rectangle(img, (400, 700), (600, 1000), (102, 178, 255), 5)
cv2.rectangle(img, (250, 450), (350, 600), (102, 178, 255), 5)
cv2.rectangle(img, (650, 450), (750, 600), (102, 178, 255), 5)

#img[0:400,:] = [135, 206, 235]

cv2.circle(img, (500, 300), 80, (255,153, 153), 3)
cv2.circle(img, (575, 850), 25, (255,153, 153), 3)
cv2.circle(img, (825, 175), 90, (255,255, 0), 3)

cv2.line(img, (200, 400), (800, 400), (255, 0, 255), 5)
cv2.line(img, (200, 400), (500, 100), (255, 0, 255), 5)
cv2.line(img, (500, 100), (800, 400), (255, 0, 255), 5)
cv2.line(img, (300, 450), (300, 600), (102, 178, 255), 5)
cv2.line(img, (250, 525), (350, 525), (102, 178, 255), 5)
cv2.line(img, (650, 525), (750, 525), (102, 178, 255), 5)
cv2.line(img, (700, 450), (700, 600), (102, 178, 255), 5)

font = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
cv2.putText(img, 'HOUSE', (50, 200),
            font, 3, (0, 255, 0), 2, cv2.LINE_AA)

plt.imshow(img) #cv2.imshow('Draw shapes',img)

plt.show() #cv2.waitKey(0)
```
