```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("images.jpeg", cv2.IMREAD_COLOR)
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(RGB_img)
```

```python
Gray_img = cv2.imread('images.jpeg', 0)
```

```python
filename ='Gray_image.jpg'
```

```python
cv2.imwrite(filename, Gray_img)
```

```python
cv2.imshow("Original_image", img)
```

```python
cv2.imshow('Gray_Scale_img', Gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture('file_example_MP4_480_1_5MG.mp4')
```

```python
if (cap.isOpened()== False):
    print("Error opening video file")
```

```python
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        cv2.imshow('Frame', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
```
