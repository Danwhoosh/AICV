```python
import cv2

#REading the image
img = cv2.imread('faces.jpg')

#Converting image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Loading the required haar-cascade xml classifier file
haar_cascade1 = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

haar_cascade2 = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

#Applying the face detection method on the grayscale image
faces_rect = haar_cascade1.detectMultiScale(gray_img, 1.1, 9)
eyes_rect = haar_cascade2.detectMultiScale(gray_img, 1.1, 9)

#Iterating through rectangles of detected faces
for (x, y, w, h) in faces_rect:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
for (x, y, w, h) in eyes_rect:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255), 2)
    
cv2.imshow('Detected faces', img)

cv2.waitKey(0)
```
